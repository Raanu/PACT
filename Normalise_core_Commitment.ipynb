```python
import graphlab
import re
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
sf = graphlab.SFrame('core_alignments_raw.csv')
```

```python
len(sf)
```

```python
sf.show()
```

```python
stakeholder_name = []
org_type = []
region = []
themes = []
cc_count = []
cc_string = []
cc_col = []
p = re.compile(',')

for row in sf:
    for cc in p.split(str(row['Core Commitment'])):
        cc_col.append(cc.strip())
        stakeholder_name.append(row['Stakeholder Name'])
        org_type.append(row['Type of Organisation'])
        region.append(row['Region'])
        themes.append(row['Themes'])
        cc_count.append(row['Core Commitment Count'])
        cc_string = row['Core Commitment']

# print len(stakeholder_name)
# print len(org_type)
# print len(region)
# print len(themes)
# print len(cc_count)
# print len(cc_col)
        
        
# print stakeholder_name
# print org_type
# print region
# print themes
# print cc_count
# print cc_string
# print cc_col

new_sf = graphlab.SFrame({'stakeholder_name': stakeholder_name, 
                         'org_type': org_type , 
                         'region': region, 
                         'themes': themes, 
                         'cc_count': cc_count, 
                         'cc_col': cc_col})
new_sf.show()
```

```python
new_sf.export_csv(filename='normalised_cc.csv',delimiter=',',line_terminator='\n',header=True,quote_level='csv.quote_NONE')
```
